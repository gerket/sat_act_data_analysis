##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
import numpy as np


def compute_sd(nums):
    
    inside_list = [(i-np.mean(nums))**2 for i in nums]
    s = np.sum(inside_list)
    
    s = s * (1/len(nums))
    s = np.sqrt(s)
    return s

sd = [compute_sd(v) for k, v in data_both.items() if type(v[0])!=str]
display(sd)
np.std(data_both['Participation_act']) #looks close enough to me
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
#Null Hypothesis - There is no difference between the SAT and ACT participation rates
#Alternative Hypothesis - There is some difference between the SAT and ACT

#Level of significance (alpha) = 0.05

#Calculate the point estimates
experimental_act = data_both['Participation_act']
control_sat = data_both['Participation_sat']

mean_act = np.mean(data_both['Participation_act'])
mean_sat = np.mean(data_both['Participation_sat'])

mean_act, mean_sat
```

```python
import scipy.stats as stats

t_stat, p_value = stats.ttest_ind(experimental_act, control_sat)
t_stat, p_value
```

```python
#Visualizing the T-Statistic

xpoints = np.linspace(-4, 4, 500)
ypoints = stats.t.pdf(xpoints, 
                      len(experimental_act) + len(control_sat) - 2,
                      0, #mean of the t-distribution
                      1 #variance of the t-distribution
                     )

fig = plt.figure(figsize=(8,5))

# Get the current "axis" out of the figure.
ax = fig.gca()

# Plot the lines using `matplotlib`'s plot function:
ax.plot(xpoints, ypoints, linewidth=3, color='darkred')

# Plot a vertical line for our measured difference in rates' t-statistic.
ax.axvline(t_stat, color='black', linestyle='--', lw=5)
ax.axvline(-t_stat, color='black', linestyle='--', lw=5)
```

The above graph is a visual representation of the t-value in this hypothesis test comparing the ACT and SAT Participation rates.

**-- Question 26 Interpretation:** The p-value of 0.00024 is less than our alpha value of 0.05. Therefore, we can say that there is enough evidence to suggest that the null hypothesis is false. Since the null hypothesis was that there is no difference between the two participation rates, this means that there is a statistically significant difference between the two participation rates.

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python

#take in list of numbers, z-value
#return list of [lower bound, upper bound] for confidence interval
def confidence_interval(nums, z=1.96):
    #z = 1.96 #95% z value
    standard_error = np.std(nums, ddof=1)/np.sqrt(len(nums))
    nums_mean = np.mean(nums)
    
    ci_lower = nums_mean - z*standard_error
    ci_upper = nums_mean + z*standard_error
    return [ci_lower, ci_upper]


act_part = confidence_interval(list(data_both["Participation_act"]))
sat_part = confidence_interval(list(data_both["Participation_sat"]))


print('95% Confidence Intervals for Participation\n')
print(f'ACT: {act_part[0]:.2f} - {act_part[1]:.2f}')
print(f'SAT: {sat_part[0]:.2f} - {sat_part[1]:.2f}')
```

**-- Question 27 Interpretation:** The confidence intervals do not overlap. That means that even if the SAT participation score was actually 5% higher and the ACT participation score was actually 5% lower, they would still not be the same and ACT would be higher than SAT. This means that the potential for your conclusions to be wrong because your samples were too far off of your true population is unlikely.
